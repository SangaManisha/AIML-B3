```python
import numpy as np

# NumPy array
my_array = np.array([1, 2, 3])

# Sine function
result_sin = np.sin(my_array)
print("NumPy sin():", result_sin)

# Natural logarithm
result_log = np.log(my_array)
print("\nNumPy log():", result_log)

# Base-2 logarithm
result_log2 = np.log2(my_array)
print("\nNumPy log2():", result_log2)

# Exponential function (e^x)
result_exp = np.exp(my_array)
print("\nNumPy exp():", result_exp)
```

## part-2:Visualization of Datasets using matplotlib,pyplots packages.

```python
import matplotlib.pyplot as plt

people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

# Scatter Plot
plt.scatter(age, height, s=weight, alpha=0.7, c='blue', label='Weight')
plt.title('Scatter Plot of Age vs Height with Weight')
plt.xlabel('Age')
plt.ylabel('Height')
plt.legend()
plt.show()

# Bar Chart
plt.bar(people, weight, color='green')
plt.title('Bar Chart of People vs Weight')
plt.xlabel('People')
plt.ylabel('Weight')
plt.show()

# Histogram
plt.hist(age, bins=5, color='orange', edgecolor='black')
plt.title('Histogram of Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
